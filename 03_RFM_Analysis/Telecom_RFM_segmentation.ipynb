### K-means Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Separate features by RFM category
recency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Recency']['Variable Name'].tolist()
frequency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Frequency']['Variable Name'].tolist()
monetary_vars = rfm_cols[rfm_cols['RFM Category'] == 'Monetary']['Variable Name'].tolist()
recency_vars
```

```python
#Cluster seperately based on the 3 dimensions

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

def optimize_clusters(df, features, rfm_type, k_range=range(3, 10)):
    subset = df[features].dropna()
    scaler = StandardScaler()
    scaled = scaler.fit_transform(subset)

    best_k = None
    best_score = -1
    scores = []

    print(f"\n--- {rfm_type} Clustering Optimization ---")
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(scaled)

        score = silhouette_score(scaled, labels)
        scores.append(score)

        counts = pd.Series(labels).value_counts().sort_index()
        print(f"K={k} | Silhouette Score: {score:.4f} | Cluster sizes: {counts.tolist()}")

        if score > best_score:
            best_score = score
            best_k = k

    print(f"✅ Best K for {rfm_type}: {best_k} with silhouette score {best_score:.4f}")
    return best_k

def cluster_and_analyze(df, features, rfm_type):
    # Filter to features that actually exist in df
    valid_features = [f for f in features if f in df.columns]
    if len(valid_features) < 2:
        print(f"⚠️ Skipping {rfm_type} clustering: Not enough valid features found.")
        return df

    subset = df[valid_features].dropna()
    if subset.empty:
        print(f"⚠️ Skipping {rfm_type} clustering: No data after dropping NA.")
        return df

    scaler = StandardScaler()
    scaled = scaler.fit_transform(subset)

    # Optimize clusters
    best_k = optimize_clusters(df, valid_features, rfm_type)

    # Final KMeans
    kmeans = KMeans(n_clusters=best_k, random_state=42)
    labels = kmeans.fit_predict(scaled)

    df.loc[subset.index, f'{rfm_type}_Cluster'] = labels

    # Churn analysis
    result = df.loc[subset.index].groupby(f'{rfm_type}_Cluster')['churn'].agg(['mean', 'count']).reset_index()
    result.columns = [f'{rfm_type}_Cluster', 'Churn_Rate', 'Count']

    print(f"\n=== {rfm_type} Clusters vs Churn ===")
    print(result)

    sns.barplot(x=f'{rfm_type}_Cluster', y='Churn_Rate', data=result)
    plt.title(f'{rfm_type} Clusters vs Churn Rate')
    plt.show()

    return df

# Get RFM variable names from rfm_cols and filter only those present in the DataFrame
recency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Recency']['Variable Name'].tolist()
frequency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Frequency']['Variable Name'].tolist()
monetary_vars = rfm_cols[rfm_cols['RFM Category'] == 'Monetary']['Variable Name'].tolist()

# Apply clustering and analysis with safety for missing columns
filtered_df = cluster_and_analyze(filtered_df, recency_vars, 'Recency')
filtered_df = cluster_and_analyze(filtered_df, frequency_vars, 'Frequency')
filtered_df = cluster_and_analyze(filtered_df, monetary_vars, 'Monetary')
```

```python
# Step 1: Evaluate average silhouette score from each clustering
from sklearn.metrics import silhouette_score

def evaluate_final_clustering(df, rfm_cols):
    cluster_scores = {}
    for rfm_type in ['Recency', 'Frequency', 'Monetary']:
        cluster_col = f'{rfm_type}_Cluster'
        features = rfm_cols[rfm_cols['RFM Category'] == rfm_type]['Variable Name'].tolist()
        valid_features = [f for f in features if f in df.columns]

        subset = df.dropna(subset=valid_features + [cluster_col])
        if subset.empty or cluster_col not in subset.columns:
            continue

        scaler = StandardScaler()
        scaled = scaler.fit_transform(subset[valid_features])
        labels = subset[cluster_col].astype(int)

        try:
            score = silhouette_score(scaled, labels)
            cluster_scores[rfm_type] = score
        except:
            cluster_scores[rfm_type] = None

    print("\n=== Final Clustering Silhouette Scores ===")
    for k, v in cluster_scores.items():
        print(f"{k}: {v:.4f}" if v else f"{k}: Unable to compute")

    return cluster_scores

# Evaluate which dimension had best clustering score
clustering_scores = evaluate_final_clustering(filtered_df, rfm_cols)

# Step 2: Save clustered dataset to Excel or CSV
filtered_df.to_csv("rfm_clustered_customers.csv", index=False)
print("\n✅ Clustered customer data saved to 'rfm_clustered_customers.csv'")
```

```python
#Cluster on the combination: unified version

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def unified_rfm_clustering(df, rfm_cols, k_range=range(3, 10)):
    # Step 1: Get all valid RFM features that exist in df
    rfm_features = rfm_cols[rfm_cols['RFM Category'].isin(['Recency', 'Frequency', 'Monetary'])]['Variable Name'].tolist()
    valid_features = [f for f in rfm_features if f in df.columns]

    if len(valid_features) < 2:
        print("❌ Not enough valid RFM features for clustering.")
        return df

    # Step 2: Prepare data
    rfm_data = df[valid_features].dropna()
    if rfm_data.empty:
        print("❌ No valid rows to cluster after dropping missing values.")
        return df

    scaler = StandardScaler()
    rfm_scaled = scaler.fit_transform(rfm_data)

    # Step 3: Find optimal K using silhouette score
    best_k = None
    best_score = -1
    print("\n--- RFM Cluster Optimization ---")
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(rfm_scaled)
        score = silhouette_score(rfm_scaled, labels)
        counts = pd.Series(labels).value_counts().sort_index()
        print(f"K={k} | Silhouette Score: {score:.4f} | Cluster sizes: {counts.tolist()}")

        if score > best_score:
            best_k = k
            best_score = score

    print(f"\n✅ Best K: {best_k} with Silhouette Score: {best_score:.4f}")

    # Step 4: Apply KMeans with best_k
    final_kmeans = KMeans(n_clusters=best_k, random_state=42)
    final_labels = final_kmeans.fit_predict(rfm_scaled)
    df.loc[rfm_data.index, 'RFM_Cluster'] = final_labels

    # Step 5: Churn analysis by cluster
    result = df.loc[rfm_data.index].groupby('RFM_Cluster')['churn'].agg(['mean', 'count']).reset_index()
    result.columns = ['RFM_Cluster', 'Churn_Rate', 'Count']
    print("\n=== Churn by RFM Cluster ===")
    print(result)

    # Step 6: Visualization
    sns.barplot(x='RFM_Cluster', y='Churn_Rate', data=result)
    plt.title("Churn Rate by Unified RFM Cluster")
    plt.ylabel("Churn Rate")
    plt.xlabel("Cluster")
    plt.show()

    return df

filtered_df = unified_rfm_clustering(filtered_df, rfm_cols)
```

### PCA-enables clusters

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def pca_rfm_clustering(df, rfm_cols, n_components=3, k_range=range(3, 10)):
    # Step 1: Select valid RFM features
    rfm_features = rfm_cols[rfm_cols['RFM Category'].isin(['Recency', 'Frequency', 'Monetary'])]['Variable Name'].tolist()
    valid_features = [f for f in rfm_features if f in df.columns]

    if len(valid_features) < 2:
        print("❌ Not enough RFM features.")
        return df

    rfm_data = df[valid_features].dropna()
    if rfm_data.empty:
        print("❌ No valid rows.")
        return df

    # Step 2: Scale and apply PCA
    scaler = StandardScaler()
    rfm_scaled = scaler.fit_transform(rfm_data)

    pca = PCA(n_components=n_components)
    rfm_pca = pca.fit_transform(rfm_scaled)

    print(f"Explained Variance Ratio by PCA components: {pca.explained_variance_ratio_}")

    # Step 3: Determine best K using silhouette
    best_k, best_score = None, -1
    print("\n--- PCA-based RFM Cluster Optimization ---")
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(rfm_pca)
        score = silhouette_score(rfm_pca, labels)
        print(f"K={k} | Silhouette Score: {score:.4f}")
        if score > best_score:
            best_score = score
            best_k = k

    print(f"\n✅ Best K: {best_k} with silhouette score {best_score:.4f}")

    # Step 4: Final clustering and analysis
    final_kmeans = KMeans(n_clusters=best_k, random_state=42)
    final_labels = final_kmeans.fit_predict(rfm_pca)
    df.loc[rfm_data.index, 'PCA_RFM_Cluster'] = final_labels

    result = df.loc[rfm_data.index].groupby('PCA_RFM_Cluster')['churn'].agg(['mean', 'count']).reset_index()
    result.columns = ['PCA_RFM_Cluster', 'Churn_Rate', 'Count']
    print("\n=== Churn by PCA-RFM Cluster ===")
    print(result)

    sns.barplot(x='PCA_RFM_Cluster', y='Churn_Rate', data=result)
    plt.title("Churn Rate by PCA-based RFM Cluster")
    plt.show()

    return df

filtered_df = pca_rfm_clustering(filtered_df, rfm_cols)
```
