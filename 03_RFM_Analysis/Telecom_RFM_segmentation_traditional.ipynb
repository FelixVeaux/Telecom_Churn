```python
#Cluster seperately based on the 3 dimensions

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

def optimize_clusters(df, features, rfm_type, k_range=range(3, 10)):
    subset = df[features].dropna()
    scaler = StandardScaler()
    scaled = scaler.fit_transform(subset)

    best_k = None
    best_score = -1
    scores = []

    print(f"\n--- {rfm_type} Clustering Optimization ---")
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(scaled)

        score = silhouette_score(scaled, labels)
        scores.append(score)

        counts = pd.Series(labels).value_counts().sort_index()
        print(f"K={k} | Silhouette Score: {score:.4f} | Cluster sizes: {counts.tolist()}")

        if score > best_score:
            best_score = score
            best_k = k

    print(f" Best K for {rfm_type}: {best_k} with silhouette score {best_score:.4f}")
    return best_k

def cluster_and_analyze(df, features, rfm_type):
    # Filter to features that actually exist in df
    valid_features = [f for f in features if f in df.columns]
    if len(valid_features) < 2:
        print(f" Skipping {rfm_type} clustering: Not enough valid features found.")
        return df

    subset = df[valid_features].dropna()
    if subset.empty:
        print(f" Skipping {rfm_type} clustering: No data after dropping NA.")
        return df

    scaler = StandardScaler()
    scaled = scaler.fit_transform(subset)

    # Optimize clusters
    best_k = optimize_clusters(df, valid_features, rfm_type)

    # Final KMeans
    kmeans = KMeans(n_clusters=best_k, random_state=42)
    labels = kmeans.fit_predict(scaled)

    df.loc[subset.index, f'{rfm_type}_Cluster'] = labels

    # Churn analysis
    result = df.loc[subset.index].groupby(f'{rfm_type}_Cluster')['churn'].agg(['mean', 'count']).reset_index()
    result.columns = [f'{rfm_type}_Cluster', 'Churn_Rate', 'Count']

    print(f"\n=== {rfm_type} Clusters vs Churn ===")
    print(result)

    sns.barplot(x=f'{rfm_type}_Cluster', y='Churn_Rate', data=result)
    plt.title(f'{rfm_type} Clusters vs Churn Rate')
    plt.show()

    return df

# Get RFM variable names from rfm_cols and filter only those present in the DataFrame
recency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Recency']['Variable Name'].tolist()
frequency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Frequency']['Variable Name'].tolist()
monetary_vars = rfm_cols[rfm_cols['RFM Category'] == 'Monetary']['Variable Name'].tolist()

# Apply clustering and analysis with safety for missing columns
filtered_df = cluster_and_analyze(filtered_df, recency_vars, 'Recency')
filtered_df = cluster_and_analyze(filtered_df, frequency_vars, 'Frequency')
filtered_df = cluster_and_analyze(filtered_df, monetary_vars, 'Monetary')
```

```python
# Step 1: Evaluate average silhouette score from each clustering
from sklearn.metrics import silhouette_score

def evaluate_final_clustering(df, rfm_cols):
    cluster_scores = {}
    for rfm_type in ['Recency', 'Frequency', 'Monetary']:
        cluster_col = f'{rfm_type}_Cluster'
        features = rfm_cols[rfm_cols['RFM Category'] == rfm_type]['Variable Name'].tolist()
        valid_features = [f for f in features if f in df.columns]

        subset = df.dropna(subset=valid_features + [cluster_col])
        if subset.empty or cluster_col not in subset.columns:
            continue

        scaler = StandardScaler()
        scaled = scaler.fit_transform(subset[valid_features])
        labels = subset[cluster_col].astype(int)

        try:
            score = silhouette_score(scaled, labels)
            cluster_scores[rfm_type] = score
        except:
            cluster_scores[rfm_type] = None

    print("\n=== Final Clustering Silhouette Scores ===")
    for k, v in cluster_scores.items():
        print(f"{k}: {v:.4f}" if v else f"{k}: Unable to compute")

    return cluster_scores

# Evaluate which dimension had best clustering score
clustering_scores = evaluate_final_clustering(filtered_df, rfm_cols)

# Step 2: Save clustered dataset to Excel or CSV
filtered_df.to_csv("rfm_clustered_customers.csv", index=False)
print("\n Clustered customer data saved to 'rfm_clustered_customers.csv'")
```

```python
#Cluster on the combination: unified version

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def unified_rfm_clustering(df, rfm_cols, k_range=range(3, 10)):
    # Step 1: Get all valid RFM features that exist in df
    rfm_features = rfm_cols[rfm_cols['RFM Category'].isin(['Recency', 'Frequency', 'Monetary'])]['Variable Name'].tolist()
    valid_features = [f for f in rfm_features if f in df.columns]

    if len(valid_features) < 2:
        print(" Not enough valid RFM features for clustering.")
        return df

    # Step 2: Prepare data
    rfm_data = df[valid_features].dropna()
    if rfm_data.empty:
        print(" No valid rows to cluster after dropping missing values.")
        return df

    scaler = StandardScaler()
    rfm_scaled = scaler.fit_transform(rfm_data)

    # Step 3: Find optimal K using silhouette score
    best_k = None
    best_score = -1
    print("\n--- RFM Cluster Optimization ---")
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(rfm_scaled)
        score = silhouette_score(rfm_scaled, labels)
        counts = pd.Series(labels).value_counts().sort_index()
        print(f"K={k} | Silhouette Score: {score:.4f} | Cluster sizes: {counts.tolist()}")

        if score > best_score:
            best_k = k
            best_score = score

    print(f"\n Best K: {best_k} with Silhouette Score: {best_score:.4f}")

    # Step 4: Apply KMeans with best_k
    final_kmeans = KMeans(n_clusters=best_k, random_state=42)
    final_labels = final_kmeans.fit_predict(rfm_scaled)
    df.loc[rfm_data.index, 'RFM_Cluster'] = final_labels

    # Step 5: Churn analysis by cluster
    result = df.loc[rfm_data.index].groupby('RFM_Cluster')['churn'].agg(['mean', 'count']).reset_index()
    result.columns = ['RFM_Cluster', 'Churn_Rate', 'Count']
    print("\n=== Churn by RFM Cluster ===")
    print(result)

    # Step 6: Visualization
    sns.barplot(x='RFM_Cluster', y='Churn_Rate', data=result)
    plt.title("Churn Rate by Unified RFM Cluster")
    plt.ylabel("Churn Rate")
    plt.xlabel("Cluster")
    plt.show()

    return df

filtered_df = unified_rfm_clustering(filtered_df, rfm_cols)
```

### PCA-enables clusters

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def pca_rfm_clustering(df, rfm_cols, n_components=3, k_range=range(3, 10)):
    # Step 1: Select valid RFM features
    rfm_features = rfm_cols[rfm_cols['RFM Category'].isin(['Recency', 'Frequency', 'Monetary'])]['Variable Name'].tolist()
    valid_features = [f for f in rfm_features if f in df.columns]

    if len(valid_features) < 2:
        print(" Not enough RFM features.")
        return df

    rfm_data = df[valid_features].dropna()
    if rfm_data.empty:
        print(" No valid rows.")
        return df

    # Step 2: Scale and apply PCA
    scaler = StandardScaler()
    rfm_scaled = scaler.fit_transform(rfm_data)

    pca = PCA(n_components=n_components)
    rfm_pca = pca.fit_transform(rfm_scaled)

    print(f"Explained Variance Ratio by PCA components: {pca.explained_variance_ratio_}")

    # Step 3: Determine best K using silhouette
    best_k, best_score = None, -1
    print("\n--- PCA-based RFM Cluster Optimization ---")
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(rfm_pca)
        score = silhouette_score(rfm_pca, labels)
        print(f"K={k} | Silhouette Score: {score:.4f}")
        if score > best_score:
            best_score = score
            best_k = k

    print(f"\n Best K: {best_k} with silhouette score {best_score:.4f}")

    # Step 4: Final clustering and analysis
    final_kmeans = KMeans(n_clusters=best_k, random_state=42)
    final_labels = final_kmeans.fit_predict(rfm_pca)
    df.loc[rfm_data.index, 'PCA_RFM_Cluster'] = final_labels

    result = df.loc[rfm_data.index].groupby('PCA_RFM_Cluster')['churn'].agg(['mean', 'count']).reset_index()
    result.columns = ['PCA_RFM_Cluster', 'Churn_Rate', 'Count']
    print("\n=== Churn by PCA-RFM Cluster ===")
    print(result)

    sns.barplot(x='PCA_RFM_Cluster', y='Churn_Rate', data=result)
    plt.title("Churn Rate by PCA-based RFM Cluster")
    plt.show()

    return df

filtered_df = pca_rfm_clustering(filtered_df, rfm_cols)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def cluster_on_all_features(df, exclude_cols=['customer_id', 'churn'], k_range=range(3, 10)):
    # Step 1: Select valid numeric features
    feature_cols = [
        col for col in df.columns
        if col not in exclude_cols and pd.api.types.is_numeric_dtype(df[col])
    ]

    if len(feature_cols) < 2:
        print(" Not enough features to cluster.")
        return df

    # Step 2: Drop NA rows and scale
    data = df[feature_cols].dropna()
    if data.empty:
        print(" No data available after dropping NA rows.")
        return df

    scaler = StandardScaler()
    scaled = scaler.fit_transform(data)

    # Step 3: Determine best K using silhouette score
    best_k = None
    best_score = -1
    print("\n--- Cluster Optimization (All Features) ---")
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(scaled)
        score = silhouette_score(scaled, labels)
        counts = pd.Series(labels).value_counts().sort_index()
        print(f"K={k} | Silhouette Score: {score:.4f} | Cluster sizes: {counts.tolist()}")

        if score > best_score:
            best_k = k
            best_score = score

    print(f"\n Best K: {best_k} with Silhouette Score: {best_score:.4f}")

    # Step 4: Final clustering
    final_kmeans = KMeans(n_clusters=best_k, random_state=42)
    final_labels = final_kmeans.fit_predict(scaled)
    df.loc[data.index, 'All_Features_Cluster'] = final_labels

    # Step 5: Churn analysis by cluster
    result = df.loc[data.index].groupby('All_Features_Cluster')['churn'].agg(['mean', 'count']).reset_index()
    result.columns = ['All_Features_Cluster', 'Churn_Rate', 'Count']
    print("\n=== Churn by Cluster (All Features) ===")
    print(result)

    # Step 6: Bar plot of churn
    plt.figure(figsize=(8, 5))
    sns.barplot(x='All_Features_Cluster', y='Churn_Rate', data=result)
    plt.title("Churn Rate by Cluster (All Features)")
    plt.ylabel("Churn Rate")
    plt.xlabel("Cluster")
    plt.tight_layout()
    plt.show()

    # Step 7: Feature averages by cluster
    cluster_feature_means = df.loc[data.index].groupby('All_Features_Cluster')[feature_cols].mean()
    cluster_feature_means_T = cluster_feature_means.T
    cluster_feature_means_T.columns = [f"Cluster_{col}" for col in cluster_feature_means_T.columns]

    variation = cluster_feature_means_T.max(axis=1) - cluster_feature_means_T.min(axis=1)
    top_variation_features = variation.sort_values(ascending=False).head(20).index

    print("\n=== Feature Averages by Cluster (Top 20 Varying) ===")
    print(cluster_feature_means_T.loc[top_variation_features])

    # Step 8: Heatmap of top varying features
    plt.figure(figsize=(12, 8))
    sns.heatmap(cluster_feature_means_T.loc[top_variation_features], annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Top 20 Feature Averages by Cluster")
    plt.ylabel("Feature")
    plt.xlabel("Cluster")
    plt.tight_layout()
    plt.show()

    return df


log_df = cluster_on_all_features(log_df)
```

### Traditional RFM

Since the silhouette scores for the unified clustering was not hight and it is not that easily interpretable, we diecided to go with the traditional approach as well.
We will assign a low medium high level to each dimension (R,F,M) adn then look at the 27 possible segments and maybe compbine some of them if they do not make much difference in terms of churn or are small groups.

```python
#First check the distributions

import matplotlib.pyplot as plt
import pandas as pd

def inspect_all_rfm_distributions(df, rfm_cols):
    rfm_types = ['Recency', 'Frequency', 'Monetary']
    summary = {}

    for rfm_type in rfm_types:
        print(f"\n=========== {rfm_type.upper()} FEATURES ===========")
        vars_ = rfm_cols[rfm_cols['RFM Category'] == rfm_type]['Variable Name'].tolist()
        rfm_summary = {}

        for feature in vars_:
            if feature not in df.columns:
                print(f"Feature {feature} not found in DataFrame.")
                continue

            series = df[feature].dropna()

            # Skip boolean and categorical columns
            if pd.api.types.is_bool_dtype(series) or pd.api.types.is_categorical_dtype(series):
                print(f"Skipping boolean/categorical feature: {feature}")
                continue

            # Convert object-type to numeric if possible
            if pd.api.types.is_object_dtype(series):
                series = pd.to_numeric(series, errors='coerce').dropna()

            # Now check if it's numeric
            if series.empty or not pd.api.types.is_numeric_dtype(series):
                print(f"Skipping non-numeric or empty feature: {feature}")
                continue

            quantiles = series.quantile([0.33, 0.66, 0.90])
            value_counts = series.value_counts(normalize=True).sort_index()

            rfm_summary[feature] = {
                'quantiles': quantiles,
                'value_counts': value_counts
            }

            print(f"\n--- Feature: {feature} ---")
            print("Quantiles (33%, 66%, 90%):")
            print(quantiles)
            print("Top 10 Value Distribution (relative):")
            print(value_counts.head(10))

            plt.figure(figsize=(6, 4))
            plt.hist(series, bins=30)
            plt.title(f'{rfm_type} Feature Distribution: {feature}')
            plt.xlabel(feature)
            plt.ylabel('Frequency')
            plt.grid(True)
            plt.tight_layout()
            plt.show()

        summary[rfm_type] = rfm_summary

    return summary



rfm_feature_summary = inspect_all_rfm_distributions(log_df, rfm_cols)
```

```python
#First make levels for each dimension and then cluster
def assign_rfm_levels(df, rfm_cols):
    # Extract feature names
    recency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Recency']['Variable Name'].tolist()
    frequency_vars = rfm_cols[rfm_cols['RFM Category'] == 'Frequency']['Variable Name'].tolist()
    monetary_vars = rfm_cols[rfm_cols['RFM Category'] == 'Monetary']['Variable Name'].tolist()

    # Use the average or single best variable from each group
    # For simplicity, use the first valid column in each category
    recency_col = next((col for col in recency_vars if col in df.columns), None)
    frequency_col = next((col for col in frequency_vars if col in df.columns), None)
    monetary_col = next((col for col in monetary_vars if col in df.columns), None)

    # R: lower is better  reverse quantiles
    df['R_score'] = pd.qcut(df[recency_col], q=3, labels=[3, 2, 1])

    # F & M: higher is better
    df['F_score'] = pd.qcut(df[frequency_col], q=3, labels=[1, 2, 3])
    df['M_score'] = pd.qcut(df[monetary_col], q=3, labels=[1, 2, 3])

    # Combine to form RFM segment
    df['RFM_Segment'] = df['R_score'].astype(str) + df['F_score'].astype(str) + df['M_score'].astype(str)

    return df

assign_rfm_levels(log_df, rfm_cols)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns


# --- Step 1: Compute standardized average scores for each RFM dimension ---
available_bins = []
for category in ['Recency', 'Frequency', 'Monetary']:
    rfm_vars = rfm_cols[rfm_cols['RFM Category'] == category]['Variable Name'].tolist()
    rfm_vars = [v for v in rfm_vars if v in log_df.columns]

    if not rfm_vars:
        continue

    # Standardize and average
    scaler = StandardScaler()
    scaled = scaler.fit_transform(log_df[rfm_vars])
    avg_score = scaled.mean(axis=1)
    key = category[0]  # R, F, M

    log_df[f'{key}_score'] = avg_score
    bin_labels = [3, 2, 1] if key == 'R' else [1, 2, 3]  # Recency: lower is better
    log_df[f'{key}_bin'] = pd.qcut(avg_score, q=3, labels=bin_labels)
    available_bins.append(f'{key}_bin')

# --- Step 2: Create RFM Segment String ---
if len(available_bins) == 3:
    log_df['RFM_Segment'] = log_df[available_bins].astype(str).agg(''.join, axis=1)
else:
    print(" Not all RFM bins are available. Found:", available_bins)
    log_df['RFM_Segment'] = np.nan

# --- Step 3: Merge Small Segments ---
min_segment_size = 100
segment_counts = log_df['RFM_Segment'].value_counts()
small_segments = segment_counts[segment_counts < min_segment_size].index.tolist()
valid_segments = segment_counts.index.tolist()

def find_similar_segment(seg, valid_segments):
    for i in range(len(seg)):
        for d in ['1', '2', '3']:
            if d != seg[i]:
                new_seg = seg[:i] + d + seg[i+1:]
                if new_seg in valid_segments and new_seg not in small_segments:
                    return new_seg
    return seg

log_df['RFM_Segment_Merged'] = log_df['RFM_Segment'].apply(
    lambda seg: find_similar_segment(seg, valid_segments) if seg in small_segments else seg
)

# --- Step 4: Label Segments (Optional, only if all R, F, M bins exist) ---
if all(col in log_df.columns for col in ['R_bin', 'F_bin', 'M_bin']):
    def label_rfm_segment(row):
        r, f, m = str(row['R_bin']), str(row['F_bin']), str(row['M_bin'])
        if (r, f, m) == ('3', '3', '3'):
            return 'Champions'
        elif r == '3' and f == '3':
            return 'Loyal Customers'
        elif r == '3' and m == '3':
            return 'Big Spenders'
        elif r == '2' and f == '3' and m == '3':
            return 'Potential Loyalists'
        elif r == '1' and f == '3' and m == '3':
            return 'At Risk Loyalists'
        elif r == '1' and f == '1' and m == '3':
            return 'Lost Big Spenders'
        elif r == '1' and f == '1' and m == '1':
            return 'Lost Customers'
        elif r == '2' and f == '2' and m == '2':
            return 'Average Customers'
        elif r == '3' and f == '1' and m == '1':
            return 'New Customers'
        else:
            return 'Other'

    log_df['RFM_Label'] = log_df.apply(label_rfm_segment, axis=1)
else:
    log_df['RFM_Label'] = 'Unlabeled'

# --- Step 5: Churn Summary by RFM Segment ---
score_cols = ['R_score', 'F_score', 'M_score']
existing_score_cols = [col for col in score_cols if col in log_df.columns]

# Base metrics
agg_dict = {
    'churn': ['count', 'mean']
}

# Add score metrics that exist
for col in existing_score_cols:
    agg_dict[col] = 'mean'

# Run aggregation
segment_churn = log_df.groupby('RFM_Segment_Merged').agg(agg_dict)
segment_churn.columns = ['Customer_Count', 'Avg_Churn'] + [f"Avg_{col}" for col in existing_score_cols]
segment_churn = segment_churn.reset_index().sort_values(by='Avg_Churn', ascending=False)

# --- Optional: Summary by Label ---
if 'RFM_Label' in log_df.columns and log_df['RFM_Label'].nunique() > 1:
    summary_by_label = log_df.groupby('RFM_Label').agg(
        Customer_Count=('churn', 'count'),
        Avg_R_score=('R_score', 'mean'),
        Avg_F_score=('F_score', 'mean'),
        Avg_M_score=('M_score', 'mean'),
        Avg_Churn=('churn', 'mean')
    ).reset_index().sort_values(by='Avg_Churn', ascending=False)

    plt.figure(figsize=(12, 6))
    sns.barplot(data=summary_by_label, x='RFM_Label', y='Avg_Churn', palette='magma')
    plt.title("Average Churn Rate by RFM Segment Label")
    plt.ylabel("Churn Rate")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Export
    summary_by_label.to_csv("rfm_labeled_segment_summary.csv", index=False)

# Export segment churn summary
segment_churn.to_csv("rfm_churn_segment_summary.csv", index=False)
```
